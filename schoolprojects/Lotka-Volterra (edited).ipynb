```python
%matplotlib inline

import matplotlib.pyplot as plt

x= []
y = []
h = []
#create three blank lists to store: x: prey population over time; y: predator population overtime; h: time-steps

x.append(float(raw_input("Input prey population: ")))
y.append(float(raw_input("Input predator population: ")))
initial = 0
h.append(initial)
final = float(raw_input("final output time: "))
time_step = float(raw_input("Input time step: "))
a, b, d, c = float(raw_input("a ")), float(raw_input("b ")), float(raw_input("d ")), float(raw_input("c "))
#Inputs: x: initial prey population; y: initial predator population; final: final output time; time_step: time-step sizes; 
#Inputs: a: prey birth rate; b: rate at which predators eat preys; c: predator death rate; 
#d: the rate at which predator population increases by consuming preys. 

k = 1 #counting variable for the x and y lists
while initial < final:
    x.append(0) 
    y.append(0) #add one new item to x and y lists
    initial  = (initial + time_step) 
    h.append(initial) #add new step to h lists (for plotting)
    x[k] = time_step*(a*x[k-1] - b*x[k-1]*y[k-1]) + x[k-1]
    y[k] = time_step*(d*x[k-1]*y[k-1] - c*y[k-1]) + y[k-1]
     #derive the function for population sizes in each subsequent step from Euler's method
    k += 1 #increment counting variable

prey = round(x[-1],3)
predator = round(y[-1],3)
#round up prey and predator population (for display)

plt.plot(h,x)
plt.plot(h,y)
plt.xlabel("Time")
plt.ylabel("Population")
plt.title("Population of prey and predator over time")
plt.show()
#Output: plot of prey and predator population over time. 

print 'Final predicted prey population: ', prey
print "Final predicted predator population: ", predator
#Output: final predicted populations of preys and predators, rounded to 3 decimal places. 
```
